```python
result_df = test.groupby('user_id')['house'].agg(lambda x: x.value_counts().index[0])
result_df = pd.DataFrame(result_df)
```

```python
house_mapping = {
    0: "house1",
    1: "house2",
    2: "house3",
    3: "house4",
    4: "house5",
    5: "house6"
}
```

```python
sub_df = pd.read_csv("../submission.csv")
sub_df
```

```python
final_df = pd.merge(sub_df, result_df, on='user_id', how='inner')
final_df['house_y'] = final_df['house_y'].map(house_mapping)
```

```python
final_df.drop(columns='house_x', inplace=True)
final_df = final_df.rename(columns={'house_y': 'house'})
final_df
```

```python
final_df.to_csv("fixed_hat3_postprocess.csv", index=False)
```
